```python
import math
# columns BMI, AGE, SUGAR
data = [
    [33.6,50,1],
    [26.6,30,0],
    [23.4,40,0],
    [43.1,67,0],
    [35.3,23,1],
    [35.9,67,1],
    [36.7,45,1],
    [25.7,46,0],
    [23.3,29,0],
    [31.0,56,1],
    [28.5,48,0],
    [34.2,52,1],
    [29.1,33,0],
    [38.0,60,1],
    [24.5,27,0],
    [40.2,65,1]
]

def euclidean_distance(x1,x2):
    return math.sqrt((x1[0]-x2[0])**2+(x1[1]-x2[1])**2)

def knn_classify(data,query,k):
    distances=[]
    for row in data:
        dist=euclidean_distance(query,row)
        distances.append((dist,row[2]))
    distances.sort(key=lambda x:x[0])
    count0=0
    count1=0
    for i in range(k):
        if distances[i][1]==0:
            count0+=1
        else:
            count1+=1
    if count1>count0:
        return 1
    else:
        return 0

query=[32,50]
k=5

result=knn_classify(data,query,k)

print("Query:",query)
print("Predicted Sugar:",result)
```

```python
import math

data = [
    [10,15,45],
    [11,6,37],
    [12,14,48],
    [7,9,33],
    [9,14,38],
    [8,12,40],
    [6,11,35],
    [15,10,50],
    [14,8,46],
    [7,12,35],
    [10,6,36],
    [13,8,44],
    [9,7,32],
    [5,8,30],
    [5,10,30]
]

def euclidean_distance(x1,x2):
    return math.sqrt((x1[0]-x2[0])**2+(x1[1]-x2[1])**2)

def knn_regression(data,query,k):
    distances=[]
    for row in data:
        dist=euclidean_distance(query,row)
        distances.append((dist,row[2]))
    distances.sort(key=lambda x:x[0])
    total=0
    for i in range(k):
        total+=distances[i][1]
    return total/k

query=[7,8]
k=5

result=knn_regression(data,query,k)

print("Query (Length, Weight):",query)
print("Predicted Cost:",result)
```
